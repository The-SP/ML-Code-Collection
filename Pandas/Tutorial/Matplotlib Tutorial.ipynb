# Pandas - Matplotlib

- Histogram
- Area plot
- Bar plot
- Line plot
- Box plot
- Scatter plot
- Hex plot
- Density and KDE plot

```python
import pandas as pd
import numpy as np
from numpy.random import randint
%matplotlib inline
```

```python
# Create a dataframe
df = pd.DataFrame(randint(0, 101, size=(25, 4)), columns=['Sun', 'Mon', 'Tues', 'Wed'])
df.head(5)
```

## Histogram

```python
# Histogram
df['Wed'].hist()
```

```python
# Histogram with smaller bins 
df['Wed'].hist(bins=50, legend=True)
```

## Area Plot

```python
# Area Plot
# df.plot(kind='area', stacked=False)
df.abs().plot(kind='area', alpha=0.5, title='Area Plot')

# Plot 2 columns
df[['Sun', 'Wed']].abs().plot(kind='area', alpha=0.5, title='Area Plot')
```

## Bar Plot

```python
# Bar plot
# df.abs().plot(kind='bar')
df.abs().plot()
```

## Line Plots

```python
# Line plot
df.plot(kind='line')
# Or, df.plot.line()

# Line width and Title
df.abs().plot(kind='line', lw=5, title="Line Plots with Line Width")

# Line Width for only few cols
df[['Mon', 'Tues']].abs().plot(kind='line')

# Size
df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(5, 3))

# legend=False
# title='Line Chart'
# alpha=0.5
```

## Box Plot

```python
df.plot(kind='box')
# Or, df.plot.box()

# figsize=(10, 5)
# cmap='jet'
# title = 'Box Plot'

# Add Grid
df.plot(kind='box', grid=True, cmap='jet')
```

## Scatter Plot

```python
# Create a dataframe
df2 = pd.DataFrame(randint(0, 101, size=(500, 4)), columns=['Sun', 'Mon', 'Tues', 'Wed'])
```

```python
# Scatter Plotar
df2.plot.scatter(x='Mon', y='Tues', title='Scatter Plot')
# Or, df.plot(kind='scatter', x='Mon', y='Tues')

df2.plot(kind='scatter', x='Mon', y='Tues', c='Wed', cmap='magma')
```

## Hex Plot

```python
df2.plot(kind='hexbin', x='Sun', y='Mon', gridsize=15)

# gridsize=10
# cmap='magma'
# title='hex'
# grid=True
```
